```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
n = 50
x = np.linspace(-4,4,n)
```

```python
K = np.exp(-0.5*np.array([(x-xi)**2 for xi in x]))
```

```python
f = np.random.multivariate_normal(np.zeros_like(x), K, 10)
```

```python
_,ax, = plt.subplots(1,2, figsize = (10,3))
ax[0].plot(x,f.T)
sns.heatmap(data=K, cmap='Blues', ax=ax[1])
plt.show()
```
