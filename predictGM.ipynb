```python
import numpy as np
import pymc3 as pm
import arviz as az
```

```python
A = np.array([1,0,1,1,0,1])
B = np.array([1,1,1,1,0,0])
```

```python
with pm.Model() as model_A_or_B:
  theta = pm.Beta('theta', 1,1)
  pm.Bernoulli('y-lik_A', p = theta, observed = A)
  pm.Bernoulli('y-lik_B', p = theta, observed = B)
  trace_A = pm.sample(1000)
  trace_B = pm.sample(1000)
  
```

```python
az.plot_trace(trace_A)
```

```python
az.plot_trace(trace_B)
```

```python
az.summary( trace_A)
```

```python
az.summary(trace_B)
```

```python
az.plot_posterior(trace_A)
```

```python
az.plot_posterior(trace_B)
```

```python
az.plot_violin(trace_A)
```

```python
az.plot_violin(trace_B)
```
